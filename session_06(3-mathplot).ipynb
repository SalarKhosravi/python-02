## Data Visualization with Matplotlib


This notebook demonstrates how to use matplotlib to create different types of charts: 
1. A linear (line) chart
2. A pie chart
3. A bar chart

Each chart represents meaningful data to understand the power of data visualization.


### 1. Linear (Line) Chart

```python

import matplotlib.pyplot as plt

# Sample data for a linear chart (x, y)
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a line chart
plt.plot(x, y, label='Linear Growth', color='blue', marker='o')

# Adding titles and labels
plt.title('Linear Growth of Data')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# Show the chart
plt.legend()
plt.show()
```

### 2. Pie Chart

```python

# Sample data for a pie chart
sizes = [15, 30, 45, 10]
labels = ['Category A', 'Category B', 'Category C', 'Category D']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# Create a pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Adding title
plt.title('Distribution of Categories')

# Show the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### 3. Bar Chart

```python

# Sample data for a bar chart
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

# Create a bar chart
plt.bar(categories, values, color='green')

# Adding titles and labels
plt.title('Category vs Values')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the bar chart
plt.show()
```
